```python
data_churn = data_churn.drop(['customerID'] , axis = 1)
```

```python
data_churn_df = pd.get_dummies(data_churn , columns = ['gender', 'Partner', 'Dependents',
       'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'], drop_first = True)
```

```python
data_churn_df.head()
```

```python
final_data = data_churn_df.applymap(lambda x : int(x) if isinstance(x , bool) else x)
```

```python
final_data.head()
```

```python
# split the data into independent and dependent

x = final_data.drop(['Churn'],axis = 1)
y = final_data['Churn']
```

```python
x.head()
```

```python
y
```

```python
# Feature scaling required 

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler_columns = ['tenure','MonthlyCharges','TotalCharges']
x[scaler_columns] = scaler.fit_transform(x[scaler_columns])
```

```python
x.head()
```

```python
# Handling imbalance data set by using SMOTE Techniques

import imblearn
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_smote , y_smote = smote.fit_resample(x , y)
print(y.value_counts())
print()
print(y_smote.value_counts())
```

```python
# splitting the data into training and testing 

from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x_smote , y_smote , test_size = 0.2 , random_state = 42)
```

```python
# pandas profiling 

from ydata_profiling import ProfileReport
reports = ProfileReport(final_data)
reports.to_file(output_file = 'output.html')
```

# Model Building

### logistic Regression



```python
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression()
logit.fit(x_train,y_train)
```

```python
# prediction

y_pred_train = logit.predict(x_train)
y_pred_test = logit.predict(x_test)
```

```python
# Evaluation metrics

from sklearn.metrics import classification_report, accuracy_score
```

```python
print(classification_report(y_train,y_pred_train))
print()
print(classification_report(y_test,y_pred_test))
```

```python
print(accuracy_score(y_train,y_pred_train))
print()
print(accuracy_score(y_test,y_pred_test))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rforest = RandomForestClassifier()
rforest.fit(x_train , y_train)
```

```python
# prediction

y_pred_rf_train = rforest.predict(x_train)
y_pred_rf_test = rforest.predict(x_test)
```

```python
print(classification_report(y_train ,y_pred_rf_train ))
print()
print(classification_report(y_test ,y_pred_rf_test ))
```

```python
print(accuracy_score(y_train ,y_pred_rf_train ))
print()
print(accuracy_score(y_test ,y_pred_rf_test ))
```
